```
# Date extraction - version 2 with neighborhood - all data

data_licenses = []
idx = 0
for account in df_final['account_number'].unique():
    idx = idx+1
    if idx%1000==0:
      print(idx)
    dfa = df_final.loc[(df_final['account_number'] == account)]
    for license in dfa['license_revised'].unique():
        dft = dfa.loc[(df_final['license_revised']==license)]
        for neigh in dft['neighborhood'].unique():
          dfn = dft.loc[(df_final['neighborhood'] == neigh)]
          start_df = pd.to_datetime(dfn['license_term_start_date'])
          end_df = pd.to_datetime(dfn['license_term_expiration_date'])
          start_term = start_df.min()
          end_term = end_df.max()
          data_licenses.append({'account_number':account,
                      'license':license,
                      'start_license':start_term,
                      'end_license':end_term,
                      'neighborhoods':neigh
                      })
```
